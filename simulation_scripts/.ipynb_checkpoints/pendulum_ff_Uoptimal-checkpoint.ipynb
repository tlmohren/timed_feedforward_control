```python
import numpy as np
import matplotlib.pyplot as plt  
from scipy.integrate import odeint 
import dynamicpendulums.phasediagram as phd
import dynamicpendulums.singlependulum as sp
from matplotlib import animation, rc   
from IPython.display import HTML
 # plot results      
import seaborn as sns
import matplotlib
```

```python
file_to_load = 'data\controlU_lookup_Q1_R100.npy'
U_bang,X,Y = np.load( file_to_load )  
print('File loaded: ' + file_to_load)
      
# ---------controller ------------   
dt = 0.001; 
tLast =3 
tInt = np.arange(0,tLast+dt, dt) ;
n_steps = np.int(tLast/dt)
print('Number of time steps: '+ str(n_steps) )

L =1.
g = -10.  
# y0 = np.array(     ) 

plot_on = True
# theta_vec= -np.arange(0.05,0.95,0.15)*np.pi
theta_vec= -np.linspace(0.05,0.95,11) *np.pi
# print(theta)

t = np.linspace(0,dt,2)  
time_sign = 1

y_list = []
u_list = []

for theta in theta_vec:
    y0 = np.array([theta,0])
    y_n = np.zeros( (n_steps,2))
    u_n= np.zeros( (n_steps,))
#     print(y0)
    y_1 = y0

    for j in range(n_steps):
        y_n[j,:] = y_1  
        closest_x =  np.argmin( np.abs(  X[0,:]-y_1[0] ) ) 
        closest_y =  np.argmin( np.abs(  Y[:,0]-y_1[1] ) ) 

        u_n[j] = U_bang[closest_y,closest_x]/2 


        dE= sp.compute_dE(y_1[0],y_1[1]  ) 
        if dE < -19.98:
            u_n[j] = 0  
        y_2 = odeint(  sp.pendulum_ode, y_1,  t, args=(u_n[j],time_sign,)  )[1]     # ode solver 
        y_2[0] = sp.wrap2periodic(y_2[0]) 

        y_1 = y_2   
        
    y_list.append(y_n) 
    u_list.append(u_n) 
```

```python
n_IC = len(theta_vec)
plt.style.use('latex_scientificPaperStyle.mplstyle')
col = sns.color_palette("hls", 5) 

t0 = np.zeros((n_IC,1) ) 
t1 = np.zeros((n_IC,1) ) 
    
colorsList = [col[2],np.array([255,255,200])/255,col[3]] 
 
font = { 'serif': ['computer modern roman']}
plt.rc('font',**font)  
contr = matplotlib.colors.ListedColormap(colorsList)

fig = plt.figure(figsize= (7,8))  
ax1 = fig.add_subplot(1, 1, 1 )

plt.pcolor(X,Y,U_bang, cmap= contr ) 
 
i_last_list = np.zeros((n_IC,),dtype = int)
i_uOn = np.zeros((n_IC,),dtype = int)
    

for j in range( n_IC):
    
    i_last = np.where(y_list[j][:,1]<-1)[0][0]
     
    i_last_list[j] = i_last 
    
    if (plot_on==True): 
        ax1.plot( y_list[j][:i_last,0], y_list[j][:i_last,1],'k')
    
    u_on = (u_list[j] >0.9) & (range(n_steps)<i_last)
    
    if len(  np.where(u_on)[0]  )> 0 : 
        i_uOn[j]  = np.where(u_on)[0][0] 
        t0[j] =   tInt[ i_uOn[j]    ]
        t1[j] =   tInt[ np.where(u_on)[0][-1]    ]
        
    if (plot_on==True): 
        ax1.plot( y_list[j][u_on,0], y_list[j][u_on,1],'r',linewidth=3) 
    
#     ax1.text(y_list[j][0,0]-0.05 ,y_list[j][0,1] -1 , j)
# ax1.set_title('Trajectories for 6 initial conditions')
```
