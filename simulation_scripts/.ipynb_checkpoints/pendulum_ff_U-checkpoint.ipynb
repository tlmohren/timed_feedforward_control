```python
import numpy as np
import matplotlib.pyplot as plt  
from scipy.integrate import odeint 
import dynamicpendulums.phasediagram as phd
import dynamicpendulums.singlependulum as sp
from matplotlib import animation, rc   
from IPython.display import HTML
 # plot results      
import seaborn as sns
import matplotlib
```

```python
file_to_load = 'data\controlU_lookup_Q1_R100.npy'
U_bang,X,Y = np.load( file_to_load )  
print('File loaded: ' + file_to_load)
       
```

```python
dth_array = np.linspace(0,8,11) 
th = -np.pi + 0.3

# ---------controller ------------   
dt = 0.001; 
tLast = 3
tInt = np.arange(0,tLast+dt, dt) ;
n_steps = np.int(tLast/dt)
print('Number of time steps: '+ str(n_steps) )
t = np.array([0,dt]) 
time_sign = 1
 
data_list = []
     
for dth in dth_array:
    y0 = np.array([th,dth]) 
    un = 0.
    y = []
    u = []
    for j in range(n_steps):
        y.append(y0)  
        u.append(un)  
        
                # this should be a lookup table 
        angle_compute = np.arctan2( y0[1],y0[0])
        
                
        ener = sp.compute_dE(y0[0],y0[1])
                    

        if (angle_compute > np.pi/2-0.2) &(angle_compute<np.pi/2+0.2) & (ener <0):
            un = 3.
        elif (angle_compute > np.pi/2-0.2) &(angle_compute<np.pi/2+0.2) & (ener >0):
            un = -3.
        else:
            un = 0.
        
        y1 = odeint(  sp.pendulum_ode, y0,  t, args=(un ,time_sign,)  )[1]     # ode solver 
        y1[0] = sp.wrap2periodic(y1[0]) 
         
        y0 = y1 

    data_list.append( [y0, y, u] ) 
```

```python
for i,list in enumerate( data_list): 
    plt.plot( np.array(list[2]) +i*5,'k' ) 
```

```python
 
data_list = []
     
for dth in dth_array:
    y0 = np.array([th,dth]) 
    un = 0.
    y = []
    u = []
    for j in range(n_steps):
        y.append(y0)  
        u.append(un)  
         
        closest_x =  np.argmin( np.abs(  X[0,:]-y0[0] ) ) 
        closest_y =  np.argmin( np.abs(  Y[:,0]-y0[1] ) ) 
        un = U_bang[closest_y,closest_x]/2  
        
        y1 = odeint(  sp.pendulum_ode, y0,  t, args=(un ,time_sign,)  )[1]     # ode solver 
        y1[0] = sp.wrap2periodic(y1[0]) 
         
        y0 = y1 

    data_list.append( [y0, y, u] ) 
        
```
